```python
import matplotlib.pyplot as plt

# Data
categories = ['Planes', 'Drones', 'Birds', 'Helicopters']
sizes = [2277, 2163, 1920, 2002]

object_sizes = ['Small', 'Medium', 'Large']
object_counts = [3071, 4008, 1283]

# Define colors
category_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
object_colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

# Plotting
plt.figure(figsize=(8, 3))

# Plot 1: Object Categories
plt.subplot(1, 2, 1)
plt.bar(categories, sizes, color=category_colors)
plt.title('(a) Object Categories')
plt.xlabel('Categories')
plt.ylabel('Number of Images')

# Space between paragraphs
plt.subplots_adjust(wspace=0.5)
# Plot 2: Object Sizes
plt.subplot(1, 2, 2)
plt.bar(object_sizes, object_counts, color=object_colors)
plt.title('(b) Object Sizes')
plt.xlabel('Sizes')
plt.ylabel('Number of Images')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
